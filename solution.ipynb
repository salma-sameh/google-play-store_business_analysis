```python
import pandas as pd 
import numpy as np
```

```python
apps_df = pd.read_csv('google-play-store.csv')
apps_df.head()
```

Data cleaning

```python
apps_df.info()
```

```python
#change the columns object data type to categorical data type 
object_columns = apps_df.select_dtypes(include='object').columns
for col in object_columns:
    apps_df[col] = apps_df[col].astype('category')
apps_df.info()
```
